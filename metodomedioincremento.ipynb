```python
#parte 1
import numpy as np
import matplotlib.pyplot as plt
N=100
phi=np.zeros(N)
u=np.zeros(N)
derivada=np.zeros(N)
delta=0.05
phi[0]=1
u[0]=0
derivada[0]=0
e=1
c1=((u[0]**2)-e)*phi[0]
derivada[1]=derivada[0]+(c1*(delta/2))

for i in range(1,N-1):
    u[i]=u[i-1]+delta
    phi[i]=phi[i-1]+(derivada[i]*delta)
    c=((u[i]**2)-e)*phi[i]
    derivada[i+1]=derivada[i]+c*(delta)
phin=np.delete(phi,N-1)
un=np.delete(u,N-1)
plt.plot(un,phin)
plt.xlim(0,3)
plt.ylim(0,1)
plt.title('Solución para armónico simple con e=1')
plt.xlabel('u')
plt.ylabel("varphi(u)")
```
